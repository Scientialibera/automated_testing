```python
# Question / Keyword Classifier
import openai
import pandas as pd
import re
import json

openai.api_key = "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"
#openai.api_version = "2023-07-01-preview"
#openai.api_type = "azure"
#openai.api_base = "https://openaitestav-01.openai.azure.com/"

functions = [
    {
        "name": "classify_query_type",
        "description": """Use to send the correct classification of the User's input. You can only return 'Follow-Up', 'Wrong Doc', 'New Doc', 'Human Contact' or 'Other'.""",
        "parameters": {
            "type": "object",
            "properties": {
                "text_type": {
                    "type": "string",
                    "description": "The Classification of User's latest input",
                    "enum": ["Follow-Up", "Wrong Doc", "New Doc", "Other", "Human Contact"]
                }
            },
            "required": ["text_type"]
        }
    }
]

def format_for_classification(convo_list):
    # Convert list of dicts to readable conversation format
    formatted_text = ""
    for entry in convo_list:
        role = entry["role"].capitalize()
        content = entry["content"]
        # Check if convo_list has only 1 entry and it's the first iteration
        if len(convo_list) == 1 and entry == convo_list[0]:
            formatted_text += f"No previous history\n\n\n\n{role} (CLASSIFY): {content}\n"
        elif role == "User" and entry == convo_list[-1]:  # if the last message is a user message
            formatted_text += f"\n\n\n\n{role} (CLASSIFY): {content}\n"
        else:
            formatted_text += f"{role}: {content}\n"
    return formatted_text

def classify_chat_intent(query_dict):
    # Initialize list to store the results
    results = []
    
    # Iterate over the dictionary items
    for convo, ground_truth in query_dict.items():
        convo_list = json.loads(convo)
        user_text = format_for_classification(convo_list)
        
        # Construct the messages to be sent to the model
        messages = [
            {"role": "system", "content": """
You are an AI NLP Expert Classifier, specialized in analyzing user inputs in a human-bot conversation. Your task is to categorize the first / latest user input into one of five categories: 'New Doc', 'Follow-Up', 'Wrong Doc', 'Human Contact', or 'Other'. The chat may include previous conversation history. If there is no history classification can't be 'Follow-Up' or 'Wrong Doc', this is the most important part you get right. User's input might explicitly indicate the desired classification (e.g., "new doc: no account").
New Doc: This category includes most inputs without history or unrelated to the previous conversation. Look for questions or keywords that diverge from the existing conversation context, typically featuring words like "what", "how to", "where to", or introducing new subjects not previously mentioned. Classify inputs as 'New Doc' if no chat history.
Follow-Up: Inputs under this category directly relate to the ongoing topic and aim to continue the dialogue. These can be questions or keywords that seek further clarification on a previous response, often containing phrases like "why did you say/write that", "what if", "let's say", "clarify", "tell me more", or echoing parts of the earlier answer.
Wrong Doc: If input contains exactly words: "Your answer is wrong". Must contain Chat History.
Human Contact: If input contains exactly words: "Support Agent", "talk to person" only
Other: This is a catch-all category for social interactions typically used in human-bot conversations, such as greetings, farewells, expressions of gratitude.
"""}, 
            {"role": "user", "content": f"Given the context above, classify the chat intent below:\n\n{user_text}"}
        ]

        # Request classification from the model
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo-16k",
            messages=messages,
            functions=functions,
            function_call={"name": "classify_query_type"},
            temperature=0
        )

        # Extract the predicted class from the response and clean it up using regex
        dirty_class = response['choices'][0]['message']['function_call']['arguments']
        predicted_class = json.loads(dirty_class).get("text_type")
        
        # Append the formatted conversation, ground truth, and predicted class to the results list
        results.append([user_text, ground_truth, predicted_class])
    
    # Convert the results list to a dataframe
    df = pd.DataFrame(results, columns=["Conversation", "Ground Truth", "Predicted Intent"])
    
    return df

df = classify_chat_intent(questions)
df
```
